```python
# PART 2: DATA CLEANING & PREPROCESSING (Basic → Medium)
```

```python
df.isnull().sum()
```

```python
df.dropna().sum()
```

```python
df.drop_duplicates()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
```

```python
df['Year'] = df['Order Date'].dt.year
```

```python
df['Revenue'] = df['Sales'] * df['Quantity']
```

```python
# PART 4: INTERMEDIATE VISUALIZATION 
```

```python
plt.plot(df['Year'], df['Sales'], color='red', label='Sales per year')
plt.plot(df['Year'], df['Profit'], color='green', label='Profit per year')
plt.title('Graph')
plt.grid(True)
plt.show()
```

```python
plt.scatter(df['Sales'], df['Profit'])
plt.show()
```

```python
plt.plot(df['Year'], df['Sales'], color='orange', marker='o', linewidth=1)
plt.title('Graph')
plt.grid(True)
plt.show()
```

```python

yearly = df.groupby('Year')[['Sales','Profit']].sum()

fig, ax = plt.subplots(2, 1, figsize=(8,6))

ax[0].plot(yearly.index, yearly['Sales'], color='red', marker='o')
ax[0].set_title('Sales per Year')
ax[0].set_ylabel('Sales')
ax[0].grid(True)

ax[1].plot(yearly.index, yearly['Profit'], color='green', marker='s')
ax[1].set_title('Profit per Year')
ax[1].set_ylabel('Profit')
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

```python

yearly = df.groupby('Year')[['Sales','Profit']].sum()

yearly.plot(kind='bar')

plt.title("Sales and Profit per Year")
plt.xlabel("Year")
plt.ylabel("Amount")

plt.xticks(rotation=45)

plt.grid(True)
plt.show()
```

```python
# PART 5: ADVANCED VISUALIZATION 
```

```python
grouped = df.groupby('Region')['Sales'].sum()

grouped.plot(kind='bar')
plt.title("Sales per Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.grid(True)
plt.show()
```

```python

grouped = df.groupby(['Category', 'Region'])['Sales'].sum().unstack()

grouped.plot(kind='bar', stacked=True, figsize=(8,6))

plt.title("Total Sales by Category and Region")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.legend(title="Region")
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%Y')

df['Month'] = df['Order Date'].dt.month

monthly_sales = df.groupby('Month')['Sales'].sum()

plt.plot(monthly_sales.index, monthly_sales.values, marker='o', color='blue', linewidth=2)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(range(1,13))
plt.grid(True)
plt.show()
```

```python
yearly_sales = df.groupby('Year')['Sales'].sum()

plt.plot(yearly_sales.index, yearly_sales.values, marker='o', color='blue')

max_year = yearly_sales.idxmax()
max_value = yearly_sales.max()

plt.annotate(f'Highest: {max_value:.0f}',
             xy=(max_year, max_value),
             xytext=(max_year, max_value+5000),
             arrowprops=dict(facecolor='red', shrink=0.05),
             ha='center')

plt.title("Sales Trend with Highest Point Highlighted")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
# PART 6: EXPERT LEVEL (ANALYSIS + VISUAL STORYTELLING)
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)

top_products.plot(kind='bar', color='skyblue', figsize=(8,6))

plt.title("Top 5 Products by Sales")
plt.xlabel("Product Name")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
segment_profit = df.groupby('Segment')['Profit'].sum()

segment_profit.plot(kind='bar', color='teal', figsize=(8,6))

plt.title("Profit by Segment")
plt.xlabel("Segment")
plt.ylabel("Total Profit")
plt.xticks(rotation=45)
plt.grid(True, axis='y')
plt.show()
```

```python
plt.hist(df['Profit'], bins=30, color='purple', edgecolor='black')
plt.title("Profit Distribution")
plt.xlabel("Profit")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(24,16))

df.groupby('Year')['Sales'].sum().plot(ax=ax[0,0], marker='o', color='blue')
ax[0,0].set_title("Sales Trend")

df.groupby('Segment')['Profit'].sum().plot(kind='bar', ax=ax[0,1], color='teal')
ax[0,1].set_title("Profit by Segment")

df.groupby(['Category','Region'])['Sales'].sum().unstack().plot(kind='bar', stacked=True, ax=ax[1,0])
ax[1,0].set_title("Sales by Category & Region")

df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5).plot(kind='bar', ax=ax[1,1], color='orange')
ax[1,1].set_title("Top 5 Products by Sales")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
